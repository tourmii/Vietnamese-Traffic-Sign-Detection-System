```python
from ultralytics import YOLO
import cv2
import os
import matplotlib.pyplot as plt
import sys


PROJECT_ROOT = "/home/s/sangdv_student/tourism_agent/Vietnamese-Traffic-Sign-Detection-System/"

image_path = "image1.png"
model = YOLO(PROJECT_ROOT + "yolov11/yolo-viet-signs/yolov8n_viet_signs/weights/last.pt")
results_list = model.predict(image_path)

results = results_list[0]

img_with_boxes = results.plot()
plt.imshow(cv2.cvtColor(img_with_boxes, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()

output_path = "result_pred.jpg"
cv2.imwrite(output_path, img_with_boxes)
```
